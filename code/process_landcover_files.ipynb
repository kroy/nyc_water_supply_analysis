```python
import os
import shutil
import re
```

```python
# landcover_area_names = ["ashokan"]
landcover_area_names = ["croton_outer", "catskill_delaware_outer"]
nlcd_landcover_dir = os.path.relpath("../datasets/landcover/nlcd_annual")

# see https://regex101.com/r/kBFU9f/1
regex = re.compile(r"(Annual_NLCD_.*_CU_C1V1)(_[^.]*)(.*)")
```

```python
for area_name in landcover_area_names:
  print("looking at", area_name)
  folder = os.path.join(nlcd_landcover_dir, area_name)
  files = os.listdir(folder)
  for f in files:
    full_path = os.path.join(folder, f)
    if os.path.isfile(full_path):
      m = regex.match(f)
      if m:
        new_fname = m.group(1) + m.group(3)
        new_path = os.path.join(folder, new_fname)
        print(f"moving {f} to {new_fname}")
        shutil.move(full_path, new_path)
```
